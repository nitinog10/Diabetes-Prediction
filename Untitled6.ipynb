## Preprocess Data

### Subtask:
Address missing values, convert binary features to boolean, and identify categorical and numerical features for further processing.

**Reasoning**:
Based on the previous step, we found a few missing values in both `train_df` and `test_df`. Given that there is only one missing value in the affected columns, it's best to remove the rows with these missing values, as imputation for such a small number might introduce noise. Additionally, some columns like `family_history_diabetes`, `hypertension_history`, and `cardiovascular_history` are binary (0.0 or 1.0) and should be converted to boolean type for better semantic representation and potential performance benefits. Finally, to prepare for encoding and scaling, I will identify all categorical and numerical features.

```python
# Drop rows with any missing values in train_df and test_df
print(f"Original train_df shape: {train_df.shape}")
train_df.dropna(inplace=True)
print(f"New train_df shape after dropping NaNs: {train_df.shape}")

print(f"\nOriginal test_df shape: {test_df.shape}")
test_df.dropna(inplace=True)
print(f"New test_df shape after dropping NaNs: {test_df.shape}")

# Convert binary columns to boolean type
binary_cols = ['family_history_diabetes', 'hypertension_history', 'cardiovascular_history', 'diagnosed_diabetes']
for col in binary_cols:
    if col in train_df.columns:
        train_df[col] = train_df[col].astype(bool)
    if col in test_df.columns:
        test_df[col] = test_df[col].astype(bool)


# Separate features and target variable
X = train_df.drop('diagnosed_diabetes', axis=1)
y = train_df['diagnosed_diabetes']

# Identify categorical and numerical features (excluding 'id')
categorical_features = X.select_dtypes(include='object').columns.tolist()
numerical_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Remove 'id' from numerical_features as it's an identifier, not a true numerical feature
if 'id' in numerical_features:
    numerical_features.remove('id')

print(f"\nCategorical features: {categorical_features}")
print(f"Numerical features: {numerical_features}")

# Display info again to confirm changes in dtypes and non-null counts
print("\n--- Training Data Info After Preprocessing ---")
train_df.info()

print("\n--- Test Data Info After Preprocessing ---")
test_df.info()
```

### Subtask:
Select and train a suitable classification model to predict the probability of 'diagnosed_diabetes' using the preprocessed training data.

**Reasoning**:
Now that the data is preprocessed, the next step is to train a machine learning model. Given that the target variable `diagnosed_diabetes` is binary, a classification model is appropriate. I will start with Logistic Regression as a robust and interpretable baseline model. I will train the model on the `X_processed_df` (features) and `y` (target) and then evaluate its performance using accuracy and AUC-ROC score to understand its predictive power.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score

# Split the training data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_processed_df, y, test_size=0.2, random_state=42, stratify=y)

# Initialize and train the Logistic Regression model
model = LogisticRegression(random_state=42, solver='liblinear', C=0.1, max_iter=1000)
model.fit(X_train, y_train)

# Make predictions on the validation set
y_pred = model.predict(X_val)
y_pred_proba = model.predict_proba(X_val)[:, 1]

# Evaluate the model
accuracy = accuracy_score(y_val, y_pred)
auc_roc = roc_auc_score(y_val, y_pred_proba)

print(f"Model Accuracy on Validation Set: {accuracy:.4f}")
print(f"Model AUC-ROC on Validation Set: {auc_roc:.4f}")
```

## Preprocess Data

### Subtask:
Apply one-hot encoding to categorical features and scaling to numerical features using a `ColumnTransformer`.

**Reasoning**:
To prepare the identified categorical and numerical features for model training, I will employ a `ColumnTransformer`. This will allow me to apply `OneHotEncoder` to the categorical features and `StandardScaler` to the numerical features. This approach ensures that all features are in a suitable format and scale for machine learning algorithms, preventing issues caused by differing scales or non-numeric representations. The transformed data will then be converted back into a DataFrame for easier handling.

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
import pandas as pd

# Store 'id' columns separately before dropping them for preprocessing
# These ids will be used for creating the submission file later
train_ids = X['id'].copy() # X still contains 'id' at this point
test_ids = test_df['id'].copy()

# Convert boolean columns to int so they can be scaled as numerical features
bool_to_int_cols = ['family_history_diabetes', 'hypertension_history', 'cardiovascular_history']
for col in bool_to_int_cols:
    if col in X.columns:
        X[col] = X[col].astype(int)
    if col in test_df.columns:
        test_df[col] = test_df[col].astype(int)

# Drop 'id' from X now, as we have saved it for submission
X = X.drop('id', axis=1)

# Identify categorical and numerical features (after converting bools to int and dropping 'id')
categorical_features = X.select_dtypes(include='object').columns.tolist()
numerical_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Create a ColumnTransformer to apply different transformations to different columns
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ],
    remainder='drop' # Explicitly drop any other columns not handled
)

# Apply the transformations to the training data
X_processed = preprocessor.fit_transform(X)

# Get feature names after all transformations
processed_feature_names = preprocessor.get_feature_names_out()

# Convert the processed training data back to a DataFrame
X_processed_df = pd.DataFrame(X_processed, columns=processed_feature_names, index=X.index)

# Apply the transformations to the test data
# Drop 'id' from test_df as well before transforming
X_test_df_for_transform = test_df.drop('id', axis=1)
X_test_processed = preprocessor.transform(X_test_df_for_transform)

# Convert the processed test data back to a DataFrame
X_test_processed_df = pd.DataFrame(X_test_processed, columns=processed_feature_names, index=test_df.index)

print("Shape of processed training data:", X_processed_df.shape)
print("Shape of processed test data:", X_test_processed_df.shape)

print("\nProcessed Training Data Head:")
display(X_processed_df.head())

print("\nProcessed Test Data Head:")
display(X_test_processed_df.head())
```
